```python
from halotools.sim_manager import CachedHaloCatalog
from halotools.mock_observables import return_xyz_formatted_array
from halotools.mock_observables import counts_in_cylinders
from __future__ import division, print_function
```

```python
halocat = CachedHaloCatalog(simname='bolplanck',version_name='halotools_v0p4',halo_finder='rockstar',redshift=0.0)
```

```python
from HOD_models_fix import decorated_hod_model
```

```python
model = decorated_hod_model()
```

```python
from processing_functions import *
```

```python
param_names = ('alpha','logM1','sigma_logM','logM0','logMmin','mean_occupation_centrals_assembias_param1','mean_occupation_satellites_assembias_param1')
```

```python
param = p19p0
model.param_dict.update(dict(zip(param_names, param)))

model.populate_mock(halocat)

pos_gals_d = return_xyz_formatted_array(*(model.mock.galaxy_table[ax] for ax in 'xyz'), \
            velocity=model.mock.galaxy_table['vz'], velocity_distortion_dimension='z', period=250.) ##redshift space distorted
points = np.array(pos_gals_d,dtype=float)

cic2_h = counts_in_cylinders(points, points, 2, 10, period=250.)
cic5_h = counts_in_cylinders(points, points, 5, 10, period=250.)

ratio_h = ((cic2_h-1)/(cic5_h-1)).astype('float')
ratio_h[~np.isfinite(ratio_h)] = 0.

plt.hist(ratio_h,bins=np.linspace(0,1,30));
```
