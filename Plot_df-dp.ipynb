```python
def plot_dfdp(locfitdict,key,i,color,label):
    l = 1
    for seed in locfitdict.keys():
        if l:
            plt.plot(locfitdict[seed][key][i],'.',c=color,label=label)
            l = 0
        else:
            plt.plot(locfitdict[seed][key][i],'.',c=color)
```

```python
keylist = ['w','d','v','c','a','r','wd','wv','wc','wa','wr','dv','dc','da','dr','vc','va','vr','ca','cr','ar',\
        'wdv','wdc','wda','wdr','wvc','wva','wvr','wca','wcr','war','dvc','dva','dvr','dca','dcr','dar',\
         'vca','vcr','var','car','wdvc','wdva','wdvr','wdca','wdcr','wdar','wvca','wvcr','wvar','wcar',\
        'dvca','dvcr','dvar','dcar','vcar','wdvca','wdvcr','wdvar','wdcar','wvcar','dvcar','wdvcar']
def bundle_locfit_mean(locfit_dict,seeds):
    locfit_bundle_mean = {}
    for k in keylist:
        temp = []
        for seed in seeds:
            temp.append(locfit_dict[seed][k])
        locfit_bundle_mean[k] = mean_of_list(temp)
    return locfit_bundle_mean
```

```python
mean_locfit_dict_19p0 = bundle_locfit_mean(locfit_dict_19p0,locfit_dict_19p0.keys())
mean_locfit_dict_19p5 = bundle_locfit_mean(locfit_dict_19p5,locfit_dict_19p5.keys())
mean_locfit_dict_20p0 = bundle_locfit_mean(locfit_dict_20p0,locfit_dict_20p0.keys())
mean_locfit_dict_20p5 = bundle_locfit_mean(locfit_dict_20p5,locfit_dict_20p5.keys())
mean_locfit_dict_21p0 = bundle_locfit_mean(locfit_dict_21p0,locfit_dict_21p0.keys())
```
